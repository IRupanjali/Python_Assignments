```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Data for line plot
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]

# a. Plot the data
plt.figure(figsize=(6, 4))
plt.plot(x, y, marker='o', linestyle='-', color='blue')

# b. Customizations
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# Student marks data
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]

# a. Bar graph
plt.figure(figsize=(6, 4))
plt.bar(students, marks, color=['red', 'green', 'blue', 'orange'])

# b. Customizations
plt.title("Marks Scored by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.ylim(0, 100)
plt.show()
```

```python
# Revenue data
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]

# Find the index of the region with highest revenue
max_index = revenue.index(max(revenue))
explode = [0.1 if i == max_index else 0 for i in range(len(revenue))]

# a & b. Pie chart with labels and highlighted largest
plt.figure(figsize=(6, 6))
plt.pie(revenue, labels=regions, autopct='%1.1f%%', explode=explode, shadow=True, startangle=90)
plt.title("Revenue Distribution by Region")
plt.axis('equal')  # Equal aspect ratio ensures pie is a circle.
plt.show()
```

```python
# Generate 1000 random integers between 1 and 100
data = np.random.randint(1, 101, size=1000)

# Histogram
plt.figure(figsize=(6, 4))
plt.hist(data, bins=10, color='purple', edgecolor='black')

plt.title("Histogram of Random Integers")
plt.xlabel("Value Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
